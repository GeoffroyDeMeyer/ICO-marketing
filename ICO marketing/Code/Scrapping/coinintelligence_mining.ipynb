```python
from requests_html import HTMLSession
import numpy as np
import csv
import pandas as pd 
import datetime
session = HTMLSession()
```

```python
page = session.get('https://www.cointelligence.com/content/ico_list/')
```

```python
import csv
with open('ICO_coinintelligence.csv', 'w', newline='') as csvfile:
    fieldnames = ['name', 'start', 'end', 'USD raised', 'USD aimed', 'integrity', 'team', 'vision', 'product', 'marketing', 'community','risk','bitcointalk','facebook','twitter', 'instagram', 'slack','reddit', 'github', 'linkedin', 'medium', 'video', 'website', 'whitepaper', 'platform', 'accepted currency', 'CEO', 'cofounder', 'CTO', 'event', 'twitter link', 'github link','instagram link']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    
    for k in range (1,26) :
        adress = 'https://www.cointelligence.com/content/ico_list/page/' + str(k) + '/'
    
        page = session.get(adress)
    
        div = page.html.find('#icoAll')
        div = div[0]
        tr = div.find('tr')
        for elmt in tr[1:] : 
            
            data = {'name' : '' , 'start' : ' ', 'end' : ' ', 'video' : 0 , 'integrity' : 5 , 'team' : 5, 'vision' : 5, 'product' : 5, 'marketing' : 'mid', 'community' : 'mid','risk' : 'mid', 'whitepaper' : 0, 'bitcointalk' : 0,'facebook' : 0,'twitter' : 0, 'instagram' : 0, 'slack' : 0,'reddit' : 0, 'github' : 0, 'linkedin' : 0, 'medium' : 0, 'platform' : 0, 'accepted currency' : 0,'website' : 0, 'USD raised' : 0, 'USD aimed' : 0, 'CEO' : ' ', 'cofounder' : ' ', 'CTO' : ' ', 'event' : 0 ,'twitter link' : ' ', 'github link' : ' '}
            div = elmt.find('a')
            div = div[0]
            site = div.attrs['href']
        
            adress = session.get(str(site))
    
            company_name = adress.html.find('.entry-title')[0].text 
            name = ''
            k = 0
            while k < len(company_name) and company_name[k] != '(' :
                name += company_name[k]
                k += 1
                
            if (k+1 == len(company_name)) :
                continue
                
            data['name'] = name
            print(name)
            
            start = adress.html.find('.ico-time.ico-start')
            if start != [] :
                start = datetime.datetime.strptime(start[0].text.replace('Start: ','').replace(' ',''),'%b%d%Y')
                data['start'] = str(start).replace(' 00:00:00','')
            
            
            end = adress.html.find('.ico-time.ico-end')
            if end != [] :
                end = datetime.datetime.strptime(end[0].text.replace('End: ','').replace(' ',''),'%b%d%Y')
                data['end'] = str(end).replace(' 00:00:00','')
            
    
            video = adress.html.find('iframe')
            if video == [0] :
                video = 0
            else :
                video =1
            data['video'] = video
    
        
            search ='.score-analysis-block.text-block.analysis-block-2' 
            div = adress.html.find(str(search))
            if div != [] :
                div = div[0]
                div = div.find('p')
                score = div[1].text
            else :
                score = 0
            data['integrity'] = score
        
            search ='.score-analysis-block.text-block.analysis-block-3' 
            div = adress.html.find(str(search))
            if div != [] :
                div = div[0]
                div = div.find('p')
                score = div[1].text
            else :
                score = 0
            data['team'] = score

            search ='.score-analysis-block.text-block.analysis-block-4' 
            div = adress.html.find(str(search))
            if div != [] :
                div = div[0]
                div = div.find('p')
                score = div[1].text
            else :
                score = 0
            data['vision'] = score
        
            search ='.score-analysis-block.text-block.analysis-block-5' 
            div = adress.html.find(str(search))
            if div != [] :
                div = div[0]
                div = div.find('p')
                score = div[1].text
            else :
                score = 0
            data['product'] = score
        
        
            
            
            div = adress.html.find('.row-label') 
            for pos, elmt in enumerate(div) :
                if pos == 0 :
                    data['marketing'] = elmt.text
                elif pos == 1 :
                    data['community'] = elmt.text
                else :
                    data['risk'] = elmt.text
                
                
            if adress.html.find('.ut-btn') != [] :
                data['whitepaper'] = 1
            else :
                data['whitepaper'] = 0
        
        
        
            div = adress.html.find('.soc-links')[0]
            div = div.find('a')
            L = []
            for elmt in div : 
                L += [elmt.attrs['href']]
            for elmt in L :
                if 'facebook' in elmt :
                    data['facebook'] = 1
                if 'instagram' in elmt :
                    data['instagram'] = 1
                    data['instagram link'] = str(elmt).replace('https://www.instagram.com/','')
                if 'bitcointalk' in elmt :
                    data['bitcointalk'] = 1
                if 'twitter' in elmt :
                    data['twitter'] = 1
                    data['twitter link'] = str(elmt).replace('https://twitter.com/','').replace('http://www.twitter.com/','').replace('?lang=en','').replace('@','').replace('/','')
                if 'linkedin' in elmt :
                    data['linkedin'] = 1
                if 'slack' in elmt :
                    data['slack'] = 1
                if 'reddit' in elmt :
                    data['reddit'] = 1
                if 'github' in elmt :
                    data['github'] = 1
                    data['github link'] = str(elmt).replace('https://github.com/','')
                if 'medium' in elmt :
                    data['medium'] = 1
        
            div = adress.html.find('.ut-list')[0]
            div = div.find('li') 
            data['platform'] = div[4].text.replace('Platform ','')
            data['accepted currency'] = div[8].text.replace('Accepted Currency ','')
            
        
            if div[9].find('a') != [] :
                data['website'] = 1
            else :
                data['website'] = 0
            
    
    
            div = adress.html.find('.numbers')
            if div != [] :
                div = div[0].text
                if div != 'N/A' :
                    div = div.replace('(','')
                    div = div.replace('%)','')
                    pourcentage = div[len(div)-2:]
                    amount_raised = div.replace(' ETH ', '')
                    if len(amount_raised) < len(div) :
                        currency = 'ETH'
                    else :
                        amount_raised = amount_raised.replace(' BTC ', '')
                        if len(amount_raised) < len(div) :
                            currency = 'BTC'
                        else :
                            amount_raised = amount_raised.replace(' USD ', '')
                            currency = 'USD'
                    amount_raised = int(amount_raised.replace(pourcentage, '').replace(',',''))

                    if currency == 'ETH' :
                        amount_raised = amount_raised * 457.45
                        div = adress.html.find('.cap-number')
                        amount_aimed = int(div[1].text.replace(currency,'').replace(',','')) * 457.45
                        amount_aimed = int(amount_aimed)
                    elif currency == 'BTC' : 
                        amount_raised = amount_raised * 6249.52
                        div = adress.html.find('.cap-number')
                        amount_aimed = int(div[1].text.replace(currency,'').replace(',','')) * 6249.52
                        amount_aimed = int(amount_aimed)
                    
                    amount_raised = int(amount_raised)
                    pourcentage = int(pourcentage)
                    
    
                    data['USD raised'] = amount_raised
                    data['USD aimed'] = amount_aimed
    
            div = adress.html.find('.people-content')
    
    
            for pos, elmt in enumerate(div) : 
                if 'CEO' in elmt.text and pos ==0:
                    CEO = ''
                    k = 0
                    while list(elmt.text)[k] != '\n' :
                        CEO += str(elmt.text[k])
                        k += 1
                    data['CEO'] = CEO
                elif 'Co-founder' in elmt.text and pos ==0:
                    Cofounder = ''
                    k = 0
                    while list(elmt.text)[k] != '\n' :
                        Cofounder += str(elmt.text[k])
                        k += 1
                    data['cofounder'] = Cofounder
                elif 'CTO' in elmt.text :
                    CTO = ''
                    k = 0
                    while list(elmt.text)[k] != '\n' :
                        CTO += str(elmt.text[k])
                        k += 1
                    data['CTO'] = CTO
        
        
            div = adress.html.find('.ev-wrapper')
            if div != [] :
                div = div[0]
                div = div.find('.ev-item')
                event = len(div)
                data['event'] = event
        
        
            writer.writerow(data)
    
    




    


```
